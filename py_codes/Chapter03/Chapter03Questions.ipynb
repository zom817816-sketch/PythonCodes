1. 结合从中序到后序的转换算法以及计算后序表达式的算法，实现直接的中序计算。在计算时，应该使用两个栈从左往右处理中序表达式标记。一个栈用于保存运算符，另一个用于保存操作数。

```python
from pythonds import Stack

def infix_evaluator(infix_expr):
    """
    中序表达式计算器
    使用两个栈从左到右处理中序表达式
    
    参数:
        infix_expr: 字符串形式的中序表达式，每个元素间用空格分隔
    
    返回:
        表达式计算结果
    """
    # 操作数栈：存储数字
    operand_stack = Stack()
    # 运算符栈：存储运算符和括号
    operator_stack = Stack()
    
    # 运算符优先级字典
    # 数字越大，优先级越高
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}
    
    # 将表达式分割成token列表
    tokens = infix_expr.split()
    
    # 遍历每个token
    for token in tokens:
        # 处理数字（包括负数）
        if token.isdigit() or (token[0] == '-' and token[1:].isdigit()): # 处理负数
            operand_stack.push(int(token))
        
        # 处理左括号：直接压入运算符栈
        elif token == '(':
            operator_stack.push(token)
        
        # 处理右括号：计算直到遇到左括号
        elif token == ')':
            # 弹出运算符并计算，直到遇到左括号
            while operator_stack.peek() != '(':
                operator = operator_stack.pop()
                operand2 = operand_stack.pop()  # 注意：先弹出的是第二个操作数
                operand1 = operand_stack.pop()
                result = apply_operator(operator, operand1, operand2)
                operand_stack.push(result)
            # 弹出左括号（不计算）
            operator_stack.pop()
        
        # 处理运算符
        else:
            # 当运算符栈不为空，且栈顶不是左括号，且栈顶运算符优先级大于等于当前运算符时
            # 先计算栈顶运算符
            while (not operator_stack.isEmpty() and 
                   operator_stack.peek() != '(' and 
                   precedence[operator_stack.peek()] >= precedence[token]):
                operator = operator_stack.pop()
                operand2 = operand_stack.pop()
                operand1 = operand_stack.pop()
                result = apply_operator(operator, operand1, operand2)
                operand_stack.push(result)
            # 将当前运算符压入栈
            operator_stack.push(token)
    
    # 处理完所有token后，计算剩余的运算符
    while not operator_stack.isEmpty():
        operator = operator_stack.pop()
        operand2 = operand_stack.pop()
        operand1 = operand_stack.pop()
        result = apply_operator(operator, operand1, operand2)
        operand_stack.push(result)
    
    # 最终结果在操作数栈顶
    return operand_stack.pop()

def apply_operator(operator, operand1, operand2):
    """
    应用运算符计算两个操作数
    
    参数:
        operator: 运算符
        operand1: 第一个操作数
        operand2: 第二个操作数
    
    返回:
        计算结果
    """
    if operator == '+':
        return operand1 + operand2
    elif operator == '-':
        return operand1 - operand2
    elif operator == '*':
        return operand1 * operand2
    elif operator == '/':
        return operand1 / operand2
    elif operator == '^':
        return operand1 ** operand2
    else:
        raise ValueError(f'Unknown operator: {operator}')

print(infix_evaluator('3 + 4 * 2'))
print(infix_evaluator('( 3 + 4 ) * 2'))
print(infix_evaluator('10 + 2 * 6'))
print(infix_evaluator('100 * 2 + 12'))
print(infix_evaluator('100 * ( 2 + 12 )'))
print(infix_evaluator('100 * ( 2 + 12 ) / 14'))
```

2. 实现一个队列，使其添加操作和移除操作的平均时间复杂度为 (1)O 。这意味着在大多数情况下，两个操作的时间复杂度都是 (1)O ，仅在一种特殊情况下，移除操作为 ( )O n 。

```python
from pythonds import Stack

class QueueWithTwoStacks:
    """
    使用两个栈实现的队列
    添加操作时间复杂度：O(1)
    移除操作平均时间复杂度：O(1)（最坏情况O(n)）
    """
    
    def __init__(self):
        """初始化队列，创建两个栈"""
        # 输入栈：用于添加元素
        self.input_stack = Stack()
        # 输出栈：用于移除元素
        self.output_stack = Stack()
    
    def enqueue(self, item):
        """
        添加元素到队列尾部
        时间复杂度：O(1)
        """
        # 直接将元素压入输入栈
        self.input_stack.push(item)
    
    def dequeue(self):
        """
        从队列头部移除并返回元素
        平均时间复杂度：O(1)
        最坏时间复杂度：O(n)（当输出栈为空时）
        """
        # 如果输出栈为空，将输入栈的所有元素倒入输出栈
        if self.output_stack.isEmpty():
            # 将输入栈的元素全部转移到输出栈
            while not self.input_stack.isEmpty():
                self.output_stack.push(self.input_stack.pop())
        
        # 如果输出栈仍为空，说明队列为空
        if self.output_stack.isEmpty():
            raise IndexError("Queue is empty")
        
        # 从输出栈弹出元素（即队列头部元素）
        return self.output_stack.pop()
    
    def isEmpty(self):
        """检查队列是否为空"""
        return self.input_stack.isEmpty() and self.output_stack.isEmpty()
    
    def size(self):
        """返回队列的大小"""
        return self.input_stack.size() + self.output_stack.size()

# 测试代码
q = QueueWithTwoStacks()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.dequeue())  # 输出: 1
print(q.dequeue())  # 输出: 2
q.enqueue(4)
print(q.dequeue())  # 输出: 3
print(q.dequeue())  # 输出: 4
print(q.isEmpty())  # 输出: True
```
