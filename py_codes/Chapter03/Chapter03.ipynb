# 栈

栈支持如下操作：
- Stack()创建一个空栈
- push(item)将一个元素添加到顶端
- pop(item)将顶部元素移除并返回
- peek()返回顶部元素但不移除
- isEmpty()检查栈是否为空
- size()返回栈中元素数目

使用Python简单实现栈

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        return self.stack.pop()

    def peek(self):
        return self.stack[-1]

    def isEmpty(self):
        return self.stack == []

    def size(self):
        return len(self.stack)
```

## 匹配括号

编写一个算法从左到右读取一个字符串，判断括号是否匹配

```python
from pythonds import Stack

def pair_checker(string: str) -> bool:
    stack = Stack()
    for s in string:
        if s == '(':
            stack.push(s)
        elif s == ')':
            if stack.isEmpty():
                return False
            else:
                stack.pop()
        else:
            continue
    if not stack.isEmpty():
        return False
    return True

string = '11(d(w(ddd)ds)'
print(pair_checker(string))
```

存在多种括号如"([{<"的括号匹配

```python
from pythonds import Stack

def pair_checker(string: str) -> bool:
    stack = Stack()
    left, right = '([{<', ')]}>'
    for s in string:
        if s in left:
            stack.push(s)
        elif s in right:
            if stack.isEmpty():
                return False
            else:
                top = stack.pop()
                if left.find(top) != right.find(s):
                    return False
        else:
            continue
    if not stack.isEmpty():
        return False
    return True

string1 = '{(d(w(ddd<)>ds)}'
print(pair_checker(string1))
string2 = '{4(11)<fes>}34'
print(pair_checker(string2))
```

## 将十进制转换为二进制

使用“除以2”算法，并使用栈保存二进制的每一位结果

```python
from pythonds import Stack

def divide_by_2(decnum: int) -> str:
    stack = Stack()
    while decnum > 0:
        rem = decnum % 2
        stack.push(rem)
        decnum = decnum // 2
    result = ''
    while not stack.isEmpty():
        result += str(stack.pop())
    return result

decnum = 36
print(divide_by_2(decnum))
print(bin(decnum)[2:])
```

修改成一个接收十进制数以及希望转换的进制基数，“除以2”变成“除以基数”

```python
from pythonds import Stack

def baseconverter(decnum: int, base: int) -> str:
    stack = Stack()
    digits = '0123456789ABCDEF'
    while decnum > 0:
        rem = decnum % base
        stack.push(digits[rem])
        decnum = decnum // base
    result = ''
    while not stack.isEmpty():
        result += stack.pop()
    return result

decnum = 86
print(baseconverter(decnum, 8))
print(baseconverter(decnum, 16))
print(oct(decnum)[2:])
print(hex(decnum)[2:])
```

## 前序、中序和后序表达式

前序表达式要求所有运算符出现在它作用的两个操作数之前，而后序表达式则相反

开发一种用于将中序表达式转换成后序表达式的算法，研究转换过程可以发现，转化时操作数的相对位置是不变的，只有运算符改变位置。在转换过程中，由于运算符右边的操作数还未出现，因此需要保存在某处；同时，运算符有不同的优先级，可能需要反转它们的保存顺序。当遇到括号时，遇到左括号将其保存，以表示接下来会遇到高优先级的运算符，等到右括号出现时才能确定其位置；当右括号出现时，便可以将运算符取出来

```python
from pythonds import Stack
import string

class BracketMisMatchError(Exception):
    def __init__(self, msg='表达式括号不匹配') -> None:
        super().__init__(msg)

def infix_to_postfix(infix_expr: str) -> str:
    stack = Stack()
    result = []
    bracket_count = 0
    prec = {}
    prec['*'] = 3
    prec['/'] = 3
    prec['+'] = 2
    prec['-'] = 2
    prec['('] = 1
    infix_list = infix_expr.split()
    for element in infix_list:
        if element in string.ascii_letters or element.isdigit():
            result.append(element)
        elif element == '(':
            stack.push(element)
            bracket_count += 1
        elif element == ')':
            top = stack.pop()
            while top != '(':
                result.append(top)
                top = stack.pop()
            bracket_count -= 1
            if bracket_count < 0:
                raise BracketMisMatchError
        else:
            while (not stack.isEmpty()) and (prec[stack.peek()] >= prec[element]):
                result.append(stack.pop())
            stack.push(element)
    if bracket_count != 0:
        raise BracketMisMatchError
    while not stack.isEmpty():
        result.append(stack.pop())
    return ' '.join(result)

infix = 'A * B + D * K * ( E + F )'
print(infix_to_postfix(infix))
```

使用栈实现后序表达式的计算，当扫描后序表达式时，需要保存操作数而不是运算符。当算法从左往右扫描后序表达式时，未遇到运算符之前，先将操作数存入栈，遇到运算符后在取出相近的两个操作数进行相应运算后存入栈。最后栈中剩下一个值

```python
from pythonds import Stack 

def postfix_eval(postix_expr: str) -> int|float: 
    operandStack = Stack() 
    tokens = postix_expr.split() # 空格分隔的后序表达式 
    for element in tokens: 
        if element == ' ': 
            continue 
        elif element.isdigit(): 
            operandStack.push(int(element)) 
        else: 
            num1 = operandStack.pop()
            num2 = operandStack.pop()
            result = do_math(element, num2, num1)
            operandStack.push(result)
    return operandStack.pop()

def do_math(op, left, right):
    if op == '+':
        return left + right
    elif op == '-':
        return left - right
    elif op == '*':
        return left * right
    elif op == '/':
        return left / right

infix = '1 * 10 + 1 * 2 * ( 22 / 2 )'
postix_expr = infix_to_postfix(infix)
print(postfix_eval(postix_expr))
```

实现一个计算器

```python
def calculate(infix_expr: str) -> int|float:
    postfix = infix_to_postfix(infix_expr)
    output = postfix_eval(postfix)
    return output
```

# 队列

队列支持以下操作：
- Queue()创建一个空队列
- enqueue(item)在队列尾端添加一个元素
- dequeue()在队列头部移除一个元素并返回
- isEmpty()检查队列是否为空
- size()返回队列元素数量

```python
from typing import Any

# 假设队列尾端位于0位置
class Queue:
    def __init__(self) -> None:
        self.queue = []
    def enqueue(self, item: Any) -> None:
        self.queue.insert(0, item)
    def dequeue(self) -> Any:
        return self.queue.pop()
    def isEmpty(self) -> bool:
        return self.queue == []
    def size(self) -> int:
        return len(self.queue)
```

## 模拟传土豆

```python
from pythonds import Queue

def hotpotato(namelist: list|tuple, num: int) -> str:
    queue = Queue()
    for name in namelist:
        queue.enqueue(name)
    while queue.size() > 1:
        for _ in range(num):
            queue.enqueue(queue.dequeue())
        queue.dequeue()
    return queue.dequeue()

namelist = ['Peter', 'Lois', 'Boris', 'Lucas', 'Hinton']
print(hotpotato(namelist, 3))
```

## 模拟打印任务
给定一个小时，实验室有10个学生。他们这一小时最多打印2次，打印页数从1到20不等。打印机每分钟只能低质量打印10页或者高质量打印5页。如何设置打印速度。

构建一个实验室模型，学生提交打印任务时添加到等待队列中。当打印机完成一个任务后会查看队列中是否有待处理的任务。学生平均等待时间是研究的问题。

根据概率学知识，一秒内提交一个打印任务的概率为1/180

```python
from pythonds import Queue
import random
from typing import Any

# 需要实现三个类：Printer Task PrintQueue

class Printer:
    def __init__(self, pagerate) -> None:
        self.pagerate = pagerate
        self.current_task = None
        self.timeremaning = 0
    def tick(self, n: int=1) -> None:
        if self.current_task is not None:
            self.timeremaning -= n
            if self.timeremaning <= 0:
                self.current_task = None
    def busy(self):
        return self.current_task is not None
    def startNext(self, new_task: Any):
        self.current_task = new_task
        self.timeremaning = new_task.get_pages() * 60 / self.pagerate

class Task:
    def __init__(self, time) -> None:
        self.timestamp = time
        self.pages = random.randint(1, 20)
    def get_pages(self) -> int:
        return self.pages
    def get_timestamp(self) -> int:
        return self.timestamp
    def waittime(self, current_time: int) -> int:
        return current_time - self.timestamp

class Simulation:
    def __init__(self, numSeconds, pagesPerMinute) -> None:
        self.printer = Printer(pagesPerMinute)
        self.print_queue = Queue()
        self.waitingtimes = []
        self.numSeconds = numSeconds
        self.pagesPerMinute = pagesPerMinute
    def newtask(self) -> bool:
        return random.randint(1, 180) == 180
    def run(self) -> None:
        for currentSecond in range(self.numSeconds):
            if self.newtask():
                task = Task(currentSecond)
                self.print_queue.enqueue(task)
            if (not self.printer.busy()) and (not self.print_queue.isEmpty()):
                new_task = self.print_queue.dequeue()
                self.printer.startNext(new_task)
                self.waitingtimes.append(new_task.waittime(currentSecond))
            self.printer.tick()
        if self.waitingtimes:
            avg_waitingtime = sum(self.waitingtimes) / len(self.waitingtimes)
        else:
            avg_waitingtime = 0.0
        print(f'Average Wait: {avg_waitingtime:.2f} seconds Remaining Tasks: {self.print_queue.size()}')

simulation = Simulation(3600, 10)
simulation.run()
```

## 基数排序器

十进制的基数排序利用1个主桶和10个数位桶。每个桶就像一个队列，根据数字的到达的先后顺序维持其中的值。初始所有数在主桶中，然后按照述职中的每一个数位来考察这些值。第一个值从主桶中移除并根据在考察的数位将其放到对应的数位桶中。一旦所有值都放在了数位桶中，便依次从0到9号数位桶中将值放回主桶。在最后一个数位被处理完，主桶中就是排好序的值。

```python
from pythonds import Queue

def radix_sort(nums: List[int]) -> List[int]: 
    if not nums:
        return nums

    max_num = max(nums) 
    max_digit = len(str(max_num)) 

    digit_buckets = [Queue() for _ in range(10)] 
    main_bucket = Queue()
    for num in nums:
        main_bucket.enqueue(num) 
    for digit in range(max_digit):
        while not main_bucket.isEmpty():
            num = main_bucket.dequeue()
            digit_buckets[num // (10 ** digit) % 10].enqueue(num)
        for bucket in digit_buckets:
            while not bucket.isEmpty():
                main_bucket.enqueue(bucket.dequeue()) 
    sorted_nums = []
    while not main_bucket.isEmpty():
        sorted_nums.append(main_bucket.dequeue())
    return sorted_nums 
```

```python
def radix_sort(nums: List[int]) -> List[int]: 
    # 处理空列表
    if not nums: 
        return nums 
    max_num = max(nums) 
    max_digit = len(str(max_num)) 

    main_bucket = nums.copy() 
    digit_buckets = [[] for _ in range(10)] 

    for digit in range(max_digit):
        # 把主桶元素分配到数位通
        for num in main_bucket:
            digit_buckets[num // (10 ** digit) % 10].append(num) 
        # 将数位桶中的元素按顺序放回主桶
        for bucket in digit_buckets:
            main_bucket.extend(bucket)
            bucket.clear() 
    
    return main_bucket
```

# 双端队列

双端队列支持一下操作：
- Deque()创建一个双端队列
- addFront(item)将一个元素添加到双端队列前端
- addRear(item)将一个元素添加到双端队列后端
- removeFront()移除前端的一个元素并返回
- removeRear()移除后端的一个元素并返回
- is_Empty()检查双端队列是否为空
- size()返回双端队列元素数目

```python
class Deque:
    def __init__(self):
        self.deque = []
    def addFront(self, item) -> None:
        self.deque.append(item)
    def addRear(self, item) -> None:
        self.deque.insert(0, item)
    def removeFront(self):
        return self.deque.pop()
    def removeRear(self):
        return self.deque.pop(0)
    def isEmpty(self):
        return self.deque == []
    def size(self):
        return len(self.deque)
```

## 回文检测器

使用双端队列存储字符串中的字符，同时在前端和后端移除元素检测是否相同

```python
from pythonds import Deque

def palchecker(string: str) -> bool:
    deque = Deque()
    for s in string:
        deque.addFront(s)
    while deque.size() > 1:
        front = deque.removeFront()
        rear = deque.removeRear()
        if front == rear:
            continue
        return False
    return True

string1 = 'toot'
string2 = 'tfoott'
print(palchecker(string1))
print(palchecker(string2))
```

# 列表

列表是元素的集合，其中一个元素都有相对于其他元素的位置，更具体的说，这种元素是无序列表

## 实现无序列表： 链表

无序列表需要维持元素之间的相对位置，但是并不需要在连续的内存空间中维持这些位置信息。为每一个元素维护一份信息，即下一个元素的位置，那么这些元素的相对位置就通过指向下一个元素的链接来表示。

需要注意的是，必须指明列表中第一个元素的位置。

```python
# Node类
class Node:
    def __init__(self, init_data) -> None:
        self.data = init_data
        self.next = None
    def get_next(self):
        return self.next
    def set_next(self, new_next):
        self.next = new_next
    def get_data(self):
        return self.data
    def set_data(self, new_data):
        self.data = new_data

# UnorderedList类
class UnorderedList:
    def __init__(self) -> None:
        self.head = None
    def is_Empty(self):
        return self.head == None
    def add(self, item):
        temp = Node(item)
        temp.set_next(self.head)
        self.head = temp
    def length(self):
        current = self.head
        count = 0
        while current is not None:
            count += 1
            current = current.get_next()
        return count
    def search(self, item):
        current = self.head
        found = False
        while current is not None and not found:
            if current.get_data() == item:
                found = True
            else:
                current = current.get_next()
        return found
    def remove(self, item):
        current = self.head
        previous = None
        found = False
        while not found:
            if current.get_data() == item:
                found = True
            else:
                previous = current
                current = current.get_next()
        if previous is None:
            self.head = current.get_next()
        else:
            previous.set_next(current.get_next())
    def append(self, item):
        temp = Node(item)
        current = self.head
        if current is None:
            self.head = temp
        else:
            while current.get_next() is not None:
                current = current.get_next()
            current.set_next(temp)
    def insert(self, i, item):
        temp = Node(item)
        current = self.head
        previous = None
        current_idx = 0
        if i == 0 or self.is_Empty():
            temp.set_next(self.head)
            self.head = temp
            return
        else:
            while current_idx < i and current is not None:
                previous = current
                current = current.get_next()
                current_idx += 1
        previous.set_next(temp)
        temp.set_next(current)
    def index(self, item):
        current = self.head
        current_idx = 0
        found = False
        while not found and current is not None:
            if current.get_data() == item:
                found = True
            else:
                current = current.get_next()
                current_idx += 1
        if found:
            return current_idx
        else:
            return -1
    def pop(self, i=None):
        if self.is_Empty():
            return None
        current = self.head
        previous = None
        current_idx = 0
        if i is None:
            while current.get_next() is not None:
                previous = current
                current = current.get_next()
            previous.set_next(None)
        else:
            if i < 0:
                return None
            if i == 0:
                self.head = current.get_next()
            else:
                while current_idx < i and current.get_next() is not None:
                    previous = current
                    current = current.get_next()
                    current_idx += 1
                if current_idx < i:
                    previous.set_next(None)
                else:
                    previous.set_next(current.get_next())
        return current.get_data()
```
